<h1> Functions </h1>
Calculus is the study of real-valued functions

$$
f:X\longrightarrow\mathbb{R}
$$

through their rates of change. To this end we construct the even and odd monomials and note their common behavior.


__Definition__: A _monomial_ of degree 
$
d
$ 
is a function
$
f:\mathbb{R}\longrightarrow\mathbb{R}
$
given by a degree
$
d
$
polynomial
$$
f(x)=a_dx^d
$$
where
$
a_d\in\mathbb{R}
$
is a fixed real parameter called the _leading order coefficent_.



```python
"""

Even and odd monomials

"""


#   Set x axis ("x line")
#   Range from a to b with x_num points
a = -1.5
b = 1.5
x = 1000
x = np.linspace(a,b,x_num)

#   Define the even and odd monomials
#   Set integers
n = 2
k = 2
#   Define the even and odd monomials associated to n and k
f_even = x**(2*n)
f_odd = x**(2*k+1)

#   plt.plot takes in the x-values, y-values and additional arguments
plt.plot(x, f_even, label='Even $x^{2n} \,,\, n=%d$' %n)
plt.plot(x, f_odd, label =r'Odd $x^{2k+1} \,,\, k = %d$' %k)

#   Additional plot options
plt.grid(True)
plt.legend()
```

```python
"""

Monomial multiplot

"""
#   Set x axis ("x line")
#   Range from a to b with x_num points
a = -1
b = 1
x_num = 1000
x = np.linspace(a,b,x_num)

#   Make a for loop to create multiple plots.
for n in range(1,4):
    #   plt.plot takes in the x-values, y-values and additional arguments
    plt.plot(x, x**(2*n), label='Even $x^{2n} \,,\, n=%d$' %n)
    plt.plot(x, x**(2*n+1), label =r'Odd $x^{2k+1} \,,\, k = %d$' %n)

plt.legend()
plt.grid(True)
```

<h1> Limits in function space </h1>

__Caution__: This section deals with limits of sequences of functions
$
\underset{n\rightarrow\infty}{\lim}f_n(x)
$
which is a much more complicated and advanced subject than the limit of a variable of a single real valued function.
$
\underset{x\rightarrow\infty}{\lim}f(x)
$


We can see that the functions
$$
f_n(x)=x^{2n}
$$
and
$$
g_k(x)=x^{2k+1}
$$
converge to something, call it
$$
f_\infty(x)=\underset{n\rightarrow\infty}{\lim}f_n(x)
$$
and
$$
g_\infty(x)=\underset{k\rightarrow\infty}{\lim}g_k(x).
$$
I am being careful by not saying that they converge to functions since
$$
f_\infty(x)=
\begin{cases}
    1 & \text{if}\, x=-1,1\\
    0 & \text{otherwise}
\end{cases}
$$
and
$$
g_\infty(x)=
\begin{cases}
    1 & \text{if}\, x=1\\
    -1 & \text{if}\, x=-1\\
    0 & \text{otherwise}
\end{cases}
$$
are discontinuous whereas
$
f_n(x)
$
and
$
g_k(x)
$
are continuous on the interval
$
[0,1]
$
for all 
$
n,k\in\mathbb{N}_{>0}
$
.
Technically
$
f_\infty(x)
$
and
$
g_\infty(x)
$
are _distributions_, that are obtained as a limit of functions.
Let's see if this is visible when we make a GigaChad lookin ass plot with 200 monomials!

```python
"""

Monomial gigaplot

"""
#   Set x axis ("x line")
#   Range from a to b with x_num points
a = -1
b = 1
x_num = 1000
x = np.linspace(a,b,x_num)

#   Make a for loop to create multiple plots.
for n in range(1,100):
    #   plt.plot takes in the x-values, y-values and additional arguments
    plt.plot(x, x**(2*n), label='Even $x^{2n} \,,\, n=%d$' %n)
    plt.plot(x, x**(2*n+1), label =r'Odd $x^{2k+1} \,,\, k = %d$' %n)

plt.text(-0.275, 0.75, '$0< n <100$', fontsize = 22)
plt.grid(True)
```

<h1> Limits of rational functions </h1>


We can find the limit 
$
L
$
of rational functions
$$
\underset{x\rightarrow\pm\infty}{\lim}\frac{f(x)}{g(x)}=L,
$$
where
$
f,g:\mathbb{R}\longrightarrow\mathbb{R}
$
, by plotting them.
We use the plotting code from the last section to do this.

```python
"""

f/g = x/(1+x^2) limit to infinity

"""



#   Set x axis ("x line")
#   Range from a to b with x_num points
a = 0
b = 1000
x_num = 100000
x_lin = np.linspace(a,b,x_num)

#   Define the function that we're interested in
f = x_lin
g = 1+x_lin**2

#   plt.plot takes in the x-values, y-values and additional arguments
plt.plot(x_lin, f/g, label=r'$\frac{x}{1+x^2}$')
#plt.plot(x_lin, f_odd, label =r'Odd $x^{2k+1} \,,\, k = %d$' %k)

#   Additional plot options
plt.grid(True)
plt.legend()
```

Thus
$$
\underset{x\rightarrow\infty}{\lim}\frac{f(x)}{g(x)}=\underset{x\rightarrow\infty}{\lim}\frac{x}{1+x^2}=0
$$

```python
"""

f/g = (x^2+1)/(2+x) limit to infinity

"""



#   Set x axis ("x line")
#   Range from a to b with x_num points
a = 0
b = 10
x_num = 100000
x = np.linspace(a,b,x_num)

#   Define the function that we're interested in
f = x**2+1
g = 2+x

#   plt.plot takes in the x-values, y-values and additional arguments
plt.plot(x_lin, f/g, label=r'$\frac{x^2+1}{2+x}$')
#plt.plot(x_lin, f_odd, label =r'Odd $x^{2k+1} \,,\, k = %d$' %k)

#   Additional plot options
plt.grid(True)
plt.legend()
```

Thus
$$
\underset{x\rightarrow\infty}{\lim}\frac{f(x)}{g(x)}=\underset{x\rightarrow\infty}{\lim}\frac{x^2+1}{x+2}=\infty
$$
